## 6. 単純ベイズ分類

---

**
単純ベイズ分類とは、各要素(説明変数)が独立に予測対象に影響を与えていると仮定して、ベイズの定理を活用し、最もその分類が発生する確率が高いものを予測とする手法。主に文章に利用され、文章の単語をベースにした分類に使われる。<br>
**

**
目的変数を$y$,説明変数を$x_1,…,x_n$とすると単純ベイズ分類器は下記の式となる。<br>
**

　

### 単純ベイズ分類器

---


**あるサンプルがどのクラスに属するかの確率式**<br>

　$P(y|x_1,…,x_n) = P(y)\prod_{i=1}^{n} P(x_i|y)$
 
**最終的にはもっとも大きな確率が割り当たるクラスに分類する。**<br>

　$y = argmaxP(y)\prod_{i=1}^{n} P(x_i|y)$
 
<br>

## 6.1 説明変数の順位付け

---
**
説明変数の中には、食用のみのデータしかない変数など、明らかに目的変数の決定に影響していない変数が存在する。<br>
ここでは、各説明変数がどの程度目的変数に影響を及ぼしているかを数値化し、単純ベイズ分類器に利用する変数を絞る。<br>
各説明変数がどの程度目的変数に影響を及ぼしているかは、言い換えると、<br>
説明変数のカテゴリよって食用である確率と毒である確率の差が大きいことを意味する。<br>
要するに、<br>
**
<br>
**
$E_e = $ある説明変数の食用確率への影響度<br>
$P_e = $ある説明変数の食用キノコ確率<br>
$P_p = $ある説明変数の毒キノコ確率<br>
**
<br>
　$E_e= \frac{P_e}{P_p}$<br>
<br>
**
で定量化することが可能である。<br>
この食用/毒関係なく評価したいため、以下の式に置き換える。<br>
**
<br>
　$E= \begin{cases}
        \frac{P_e}{P_p} & (P_e >= P_p) \\
        \frac{P_p}{P_e} & (P_p > P_e) \\
 \end{cases}$
<br>
<br>
**
また、各特徴カテゴリの説明変数内におけるデータ数の割合も考慮に入れたいため、<br>
最終的に目的変数への影響度を以下の式で算出する。<br>
**
<br>
　$E= \sum_i P_i *
 \begin{cases}
        \frac{P_e}{P_p} & (P_e >= P_p) \\
        \frac{P_p}{P_e} & (P_p > P_e) \\
 \end{cases}$
<br>
<br>
**
各説明変数ごとに影響度を計算し、影響度順に表示する。
**

　


```python
%%time
dict_score = {}
for col, df in dict_p.items():
    df_t = df.T.copy()
    score_col = 0
    for ind, row in df_t.iterrows():
        if ind == 'All':
            continue
        score = 0
        p_e = row['e']
        p_p = row['p']
        if p_e > p_p:
            if p_p != 0:
                score = p_e / p_p 
            else:
                score = 0
        else:
            if p_e != 0:
                score = p_p / p_e
            else:
                score = 0
            
        score_col += score * row['All']
    
    dict_score[col] = [score_col]
df_score = pd.DataFrame(dict_score, index=['score']).T.sort_values(by=['score'], ascending=False)
key_values = [ind for ind in df_score.index]
display(df_score)
df_score.plot.bar()
plt.show()
    
    
```

### <span style="color:red">'odor(におい)', 'spore-print-color(胞子紋の色)'</span>の影響が大きい。
### 単純ベイズ分類器の説明変数にこの２変数を利用する。


　
 

## 6.2 モデル定義

---


```python
# 事前確率を学習する関数
def priori_probability(X_train, y_train, label, columns):
    
    data_size = len(X_train)
    
    X_train = pd.DataFrame(X_train, columns=columns)
    y_train = pd.DataFrame(y_train, columns=[label])
    
    # 事前確率
    dict_prip = {}
    for column in columns:
        df = pd.crosstab(index=y_train[label], columns=X_train[column],margins=True) / data_size
        if column == 'stalk-root':
            # 欠測値は食用キノコ確率と毒キノコ確率を均等に割り当て、目的変数に影響を与えないようにする
            df['?'] = [df['?']['All']/2, df['?']['All']/2, df['?']['All']]

        dict_prip[column] = df.copy()
    
    return dict_prip
    
```

```python
# 単純ベイズ分類器
def basian_model(X_test, prip, label, labels, columns):
    
    X_test = pd.DataFrame(X_test, columns=columns)
    
    W = {}
    for column in columns:
        for name in labels:
            W[name + column] = dict(prip[column].T[name])
  
    # 初期化
    temp = pd.DataFrame({label: [labels[0] for _ in range(len(X_test))]})
    for name in labels:
        temp['p_' + name] = prip[columns[0]]['All'][name]
        
    for column in columns:
        for name in labels:
            temp['p_' + name] *= X_test[column].map(W[name + column])
                
    temp[label] = temp[label].where(temp['p_' + labels[0]] >= temp['p_' + labels[1]], labels[1])
    
    return temp[label].values
    
```

## 6.3 交差検証

---


```python
def cross_validation(label, labels, columns):

    y = df_data[label].values
    X = df_data[columns].values

    from sklearn.model_selection import KFold

    kf = KFold(n_splits=5, random_state=1234, shuffle=True)
    kf.get_n_splits(X, y)

    df_result = pd.DataFrame()

    for train_index, test_index in kf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        prip = priori_probability(X_train, y_train, label, columns)
        y_pred = basian_model(X_test, prip, label, labels, columns)
        df = pd.DataFrame({"y_test":y_test, "y_pred":y_pred})
        df_result = pd.concat([df_result, df], axis=0)

    y_test = df_result["y_test"]
    y_pred = df_result["y_pred"]
    
    return y_test, y_pred
```

```python
%%time

# 目的変数名
label = 'class'
# 目的変数
labels = ['e','p']
# 利用する説明変数名
columns = ['odor', 'spore-print-color']

y_test, y_pred = cross_validation(label, labels, columns)
```

## 6.4 評価

---


```python
def evaluation(columns, y_test, y_pred):
    # 混同行列
    df = pd.DataFrame(confusion_matrix(y_true=y_test, y_pred=y_pred), index=[1,0], columns=[1,0])
    sns.heatmap(df, annot=True, cmap='Blues', fmt='4d')
    plt.show()

    total = df[0][0] + df[0][1] + df[1][0] + df[1][1]
    TruePositive = df[0][0]
    FalseNegative = df[0][1]
    FalsePositive = df[1][0]
    TrueNegative = df[1][1]

    accuracy = (TruePositive + TrueNegative) / total
    recall = (TruePositive) / (TruePositive + FalsePositive)
    precision = (TruePositive) / (TruePositive + FalseNegative)

    print('Columns:[')
    for column in columns:
        print('         \'' + column + '\'')
    print('        ]')
    print()
    print('Total\t\t:' + str(total))
    print('TruePositive\t:' + str(TruePositive))
    print('FalseNegative\t:' + str(FalseNegative))
    print('FalsePositive\t:' + str(FalsePositive))
    print('TrueNegative\t:' + str(TrueNegative))
    print()
    print('正解率\t：{0:.3f}%'.format(accuracy * 100))
    print('再現率\t：{0:.3f}%'.format(recall * 100))
    print('適合率\t：{0:.3f}%'.format(precision * 100))
    print('F1\t：{0:.3f}%'.format(2*precision*recall/(precision+recall) * 100))
    
    return accuracy
    
    
```

```python
evaluation(columns, y_test, y_pred)
```

## 正解率<span style="color:red">98.9%</span>達成！！

　


## 【おまけ】説明変数を増やして最大の正解率を探索する！

　


```python
accuracies = []
for i in range(1, len(key_values)+1):
    
    # 目的変数名
    label = 'class'
    # 目的変数
    labels = ['e','p']
    # 説明変数
    columns = key_values[0:i]
    
    # 交差検証
    y_test, y_pred = cross_validation(label, labels, columns)
    # 評価
    accuracy = evaluation(columns, y_test, y_pred)
    accuracies.append(accuracy)
```

```python

#print list_name.index(max(list_name))
column_num = accuracies.index(max(accuracies))
print('Column_num:' + str(column_num))
print('Columns:[')
for column in columns[0:column_num]:
    print('         \'' + column + '\'')
print('        ]')
print()
print('最大正解率:{0:.3f}%'.format(max(accuracies) * 100))
df_acc = pd.DataFrame(accuracies, index=[i for i in range(1, len(key_values)+1)], columns=['accuracy'])

plt.plot(df_acc.index, df_acc['accuracy'])
plt.show()
```

## 説明変数１７個を選択した際に正解率は最大となり、
## 最大正解率は<span style="color:red">99.95%</span>となる。
